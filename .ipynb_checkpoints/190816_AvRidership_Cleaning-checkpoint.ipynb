```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

%matplotlib inline
```

# Build basic dataframe of subway ridership

```python
# Import and clean data

df_ridership = pd.read_csv('data\Av_Weekday_Ridership.csv', encoding='latin1')
df_ridership.dropna(inplace=True)
df_ridership.reset_index(drop=True, inplace=True)
df_ridership.drop(columns=['2018 Rank'], inplace=True)
```

```python
# Convert numeric strings to numbers

def convert_num(item):
    try:
        item = int(item.replace(',', ''))
    except:
        item = item
    return item
```

```python
# Convert percentage strings to numbers

def convert_percent(item):
    try:
        item = float(item.replace('%', ''))
    except:
        item = item
    return item
```

```python
# Apply changes

df_ridership['2013'] = df_ridership['2013'].apply(convert_num)
df_ridership['2014'] = df_ridership['2014'].apply(convert_num)
df_ridership['2015'] = df_ridership['2015'].apply(convert_num)
df_ridership['2016'] = df_ridership['2016'].apply(convert_num)
df_ridership['2017'] = df_ridership['2017'].apply(convert_num)
df_ridership['2018'] = df_ridership['2018'].apply(convert_num)
```

```python
# Ensure all train names are strings
df_ridership['Train'] = df_ridership['Train'].apply(lambda x: str(x))
```

```python
# Create new columns

df_ridership['2018 Change (#)'] = df_ridership['2018'] - df_ridership['2017']
df_ridership['Total Change (#)'] = df_ridership['2018'] - df_ridership['2013']

# Percentages go to infinity if denominator is 0. Set max bound at 100%
df_ridership['2018 Change (%)'] = round(((df_ridership['2018'] - df_ridership['2017']) / df_ridership['2017']) * 100, 2)
df_ridership['2018 Change (%)'] = [x if x < 100 else 100 for x in df_ridership['2018 Change (%)']]

df_ridership['Total Change (%)'] = round(((df_ridership['2018'] - df_ridership['2013']) / df_ridership['2013']) * 100, 2)
df_ridership['Total Change (%)'] = [x if x < 100 else 100 for x in df_ridership['Total Change (%)']]
```

```python
df_ridership.sort_values(by='Total Change (%)', ascending=False)
```

# Pull out station names to create manual lists

```python
line = '6'

df_names = df_ridership[df_ridership['Train'].str.contains(line)].copy()
df_names['Train'] = line
drop_cols = ['2013', '2014', '2015', '2016', '2017', '2018', '2018 Change (#)', 
             '2018 Change (%)', 'Total Change (#)', 'Total Change (%)']

df_names.drop(columns=drop_cols, inplace=True)
df_names.reset_index(inplace=True, drop=True)
df_names.to_csv('data/stations_ordered/names_export.csv')
```

# Read in manual list of stations and create plots

```python
#Choose subway line and value to plot
line = '6'
column = 'Total Change (%)'
```

```python
df_names = pd.read_csv('data/stations_ordered/names_ordered.csv')
names = df_names[df_names['Train'].astype('str') == line]['Station'].tolist()
```

```python
x_values = [x for x in range(len(names))]
y_values = []

for name in names:
    value = df_ridership[df_ridership['Station'] == name][column].tolist()[0]
    y_values.append(value)
```

```python
#Get colors
df_colors = pd.read_csv('data\subway_colors.csv')
line_color = df_colors[df_colors['Line']==line]['Color'].tolist()[0]
```

```python
#Interpolate values to create smooth plot
length = len(x_values)-1
xnew = np.linspace(0, length, num=(length*50), endpoint=True)
func = interp1d(x_values, y_values, kind='cubic')
```

```python
#PLOT THAT HOTNESS!!!!
fig, ax = plt.subplots(figsize=(20,10))

plt.plot(xnew, func(xnew), linewidth=3, color=line_color)
plt.plot(x_values, y_values, linestyle='None', 
         marker='o', markersize=9, markerfacecolor='w', markeredgecolor=line_color, markeredgewidth=3, color=line_color)

plt.hlines(0, min(x_values)-2, max(x_values)+2, colors='k', linestyles='solid')

ax.set_facecolor('#F2F1F0')

plt.xlabel("Station")
plt.xticks(x_values, names, rotation='vertical')
plt.ylabel(column)
plt.title(f'Increase / Decrease in Ridership Along the {line} line')
plt.grid(True)

plt.savefig(f'{line}.png')
```

# Compare Train Lines

```python
names = ['A', 'C', 'E', 'B', 'D', 'F', 'M', 'G', 'J', 'Z', 'L', 'N', 'Q', 'R', 'W', 'S', '1', '2', '3', '4', '5', '6', '7']
```

```python
#Get colors
df_colors = pd.read_csv('data\subway_colors.csv')

colors = []
for name in names:
    colors.append(df_colors[df_colors['Line'] == name]['Color'].tolist()[0])
```

```python
# Create list of values (total inc/dec in ridership and average % inc/dec in ridership) for each train
total_inc = []
mean_inc = []

for name in names:
    total_increase = df_ridership[df_ridership['Train'].str.contains(name)]['Total Change (#)'].sum()
    total_ridership = df_ridership[df_ridership['Train'].str.contains(name)]['2018'].sum()
    
    total_inc.append(total_increase)
    mean_inc.append(round(total_increase/total_ridership, 2))
```

```python
y_values = mean_inc
x_values = [x for x in range(len(names))]

fig, ax = plt.subplots(figsize=(10,5))

for i in range(len(y_values)):
    plt.plot(x_values[i], y_values[i], linestyle='None', 
             marker='o', markersize=9, markerfacecolor='w', markeredgecolor=colors[i], markeredgewidth=3)

plt.grid(True)
ax.set_facecolor('#F2F1F0')
plt.hlines(0, min(x_values)-2, max(x_values)+2, colors='k', linestyles='solid', linewidth=1)
plt.xticks(x_values, names);
```

```python
y_values = total_inc
x_values = [x for x in range(len(names))]

fig, ax = plt.subplots(figsize=(10,5))

for i in range(len(y_values)):
    plt.plot(x_values[i], y_values[i], linestyle='None', 
             marker='o', markersize=9, markerfacecolor='w', markeredgecolor=colors[i], markeredgewidth=3)

plt.grid(True)
ax.set_facecolor('#F2F1F0')
plt.hlines(0, min(x_values)-2, max(x_values)+2, colors='k', linestyles='solid', linewidth=1)
plt.xticks(x_values, names);
```
