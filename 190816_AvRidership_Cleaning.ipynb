```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Import and clean data

df_ridership = pd.read_csv('data\Av_Weekday_Ridership.csv', encoding='latin1')
df_ridership.dropna(inplace=True)
df_ridership.reset_index(drop=True, inplace=True)
df_ridership.drop(columns=['2018 Rank'], inplace=True)
```

```python
# Convert numeric strings to numbers

def convert_num(item):
    try:
        item = int(item.replace(',', ''))
    except:
        item = item
    return item
```

```python
# Convert percentage strings to numbers

def convert_percent(item):
    try:
        item = float(item.replace('%', ''))
    except:
        item = item
    return item
```

```python
# Apply changes

df_ridership['2013'] = df_ridership['2013'].apply(convert_num)
df_ridership['2014'] = df_ridership['2014'].apply(convert_num)
df_ridership['2015'] = df_ridership['2015'].apply(convert_num)
df_ridership['2016'] = df_ridership['2016'].apply(convert_num)
df_ridership['2017'] = df_ridership['2017'].apply(convert_num)
df_ridership['2018'] = df_ridership['2018'].apply(convert_num)
```

```python
# Ensure all train names are strings
df_ridership['Train'] = df_ridership['Train'].apply(lambda x: str(x))
```

```python
# Create new columns

df_ridership['2018 Change (#)'] = df_ridership['2018'] - df_ridership['2017']
df_ridership['2018 Change (%)'] = round(((df_ridership['2018'] - df_ridership['2017']) / df_ridership['2017']) * 100, 2)
df_ridership['Total Change (#)'] = df_ridership['2018'] - df_ridership['2013']
df_ridership['Total Change (%)'] = round(((df_ridership['2018'] - df_ridership['2013']) / df_ridership['2013']) * 100, 2)
```
